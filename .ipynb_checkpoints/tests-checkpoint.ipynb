```python
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import os
```

```python
save_dir = 'runs\\detect\\predict'

image_names = ['img_00000021.jpg','img2.jpg']

current_path = os.getcwd()
image_path = [current_path + '\\datasets\\img\\Abstract_Buttoned_Top\\' + image_names[0], current_path + '\\datasets\\predict\\' + image_names[1]]
```

```python
def show_image(image_path):
    # Ensure the image file exists before trying to display it
    if not os.path.exists(image_path):
        print(f"Image not found: {image_path}")
        return
    
    image = mpimg.imread(image_path)
    plt.imshow(image)
    plt.axis('off')  # Turn off the axis
    plt.show()
```

```python
model = YOLO('yolov8n.pt')
```

```python
# Run batched inference on a list of images
results = model.predict(source=image_path, save=True)
```

```python
for image_name in image_names:
    # Construct the path for the saved annotated image
    # The exact path depends on where the model saves the annotated images
    annotated_image_path = os.path.join(save_dir, image_name)
    show_image(annotated_image_path)
```
