```python
import pandas as pd
from tabulate import tabulate
```

```python
df = pd.read_csv("labels/Anno_coarse/list_category_img.txt", delim_whitespace=True)
df.head()
```

```python
for i in range(50):
    if not df[df.eq(str(i)).any(axis=1)].empty :
        print(tabulate(df[df.eq(str(i)).any(axis=1)].head(1), headers='keys', tablefmt='psql'))
```
